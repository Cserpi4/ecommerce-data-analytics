```python
import pandas as pd

df = pd.read_csv("../data/raw/Online-eCommerce.csv")

df = df.dropna()
```

```python
orders = (
    df.groupby("Order_Number", as_index=False)
      .agg(
          order_date=("Order_Date", "min"),
          order_total=("Total_Sales", "sum"),
          items_count=("Product", "count"),
          status=("Status", "last")
      )
)
```

```python
orders.head()
orders.shape
```

```python
orders["order_date"] = pd.to_datetime(orders["order_date"], dayfirst=True)
```

```python
orders["month"] = orders["order_date"].dt.to_period("M").astype(str)
```

```python
orders[["order_date", "month"]].head()
```

```python
orders.to_csv("../data/processed/orders.csv", index=False)
```
