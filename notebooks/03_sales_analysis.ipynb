```python
import pandas as pd

orders = pd.read_csv("../data/processed/orders.csv")
orders.head()
```

```python
monthly = (
    orders.groupby("month", as_index=False)
    .agg(
        revenue=("order_total", "sum"),
        aov=("order_total", "mean"),
        items_per_order=("items_count", "mean"),
        orders=("Order_Number", "count")
    )
)
```

```python
monthly.head()
```

```python
monthly.to_csv("../data/processed/monthly_metrics.csv", index=False)
```
